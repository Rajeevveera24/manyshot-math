```python
import json

# Load the dataset from MATH500
with open('../datasets/MATH500/dataset.json', 'r') as file:
    math500_data = json.load(file)

# Load the dataset from MATH
with open('../datasets/MATH/test/dataset.json', 'r') as file:
    math_data = json.load(file)

# Extract questions from both datasets
math500_questions = math500_data['question']
math_questions = math_data['question']

# Find corresponding indexes
corresponding_indexes = []
for question in math500_questions:
    if question in math_questions:
        corresponding_indexes.append(math_questions.index(question))

print(corresponding_indexes)
```

```python
with open('../datasets/MATH500inMATH-idxs.json', 'w') as file:
    json.dump(corresponding_indexes, file)
```
