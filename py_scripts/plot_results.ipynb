```python
import matplotlib.pyplot as plt

n = 8

shots = [4, 5, 10, 25, 50, 75, 100, 125]
unsupervised_accuracies = [32.82, 33.75, 37.46, 41.80, 43.34, 45.2, 44.58, 41.49]
supervised_accuracies = [31.89, 36.84, 39.63, 40.87, 37.77, 35.91, 32.51, 27.5]
reinforced_accuracies = [38.08, 40.25, 40.87, 39.32, 44.58, 48.30, 38.08, 28.48]
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(32, 18))
plt.bar([i-0.3 for i in range(n)], unsupervised_accuracies, 0.3, label="Unsupervised ICL", color="lightblue")
plt.bar([i for i in range(n)], supervised_accuracies, 0.3, label="Supervised ICL", color="magenta")
plt.bar([i+0.3 for i in range(n)], reinforced_accuracies, 0.3, label="Synthetic ICL", color="orange")
plt.xticks(list(range(n)), shots, fontsize=40)
plt.yticks(fontsize=40)
plt.legend(frameon=False, fontsize=40)
plt.tight_layout()
plt.box(False)
plt.grid(axis='y')
plt.show()
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(32, 18))
plt.plot(supervised_accuracies, linewidth=20, label="Supervised ICL", color="cornflowerblue")
plt.plot(unsupervised_accuracies, linewidth=20, label="Unsupervised ICL", color="yellow")
plt.plot(reinforced_accuracies, linewidth=20, label="Synthetic ICL", color="red")
plt.xticks(list(range(n)), shots, fontsize=75)
plt.yticks(fontsize=75)
# plt.legend(frameon=False, fontsize=40)
plt.tight_layout()
plt.box(False)
plt.grid()
plt.show()
```
